## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
x_cols=attrition_df.select_dtypes(include=['number']).columns.tolist()
x_cols


# Create X_df using your selected columns
X_df=attrition_df[x_cols]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
"""I only selected int cols for X_df"""
```

```python
# Create a StandardScaler
scaler=StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled=scaler.fit_transform(X_train)


# Scale the training and testing data

X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
dep_encoder=OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
dep_encoded = dep_encoder.fit_transform(y_df[['Department']])
dep_columns = dep_encoder.get_feature_names_out(['Department'])
# Create two new variables by applying the encoder
# to the training and testing data
y_train_dep_encoded = dep_encoder.transform(y_train[['Department']])
y_test_dep_encoded = dep_encoder.transform(y_test[['Department']])
dep_encoded
```

```python
# Create a OneHotEncoder for the Attrition column
att_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
att_encoded=att_encoder.fit_transform(y_df[['Attrition']])
att_columns=att_encoder.get_feature_names_out(['Attrition'])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att_encoded = att_encoder.transform(y_train[['Attrition']])
y_test_att_encoded = att_encoder.transform(y_test[['Attrition']])
att_encoded
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_cols=X_train_scaled.shape[1]


# Create the input layer
input_layer = layers.Input(shape=(num_cols,), name='input_layer')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dep_hidden=layers.Dense(units=64,activation='relu')(shared_layer2)

# Create the output layer
dep_output = layers.Dense(3, activation='softmax', name='dep_output')(dep_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
att_hidden=layers.Dense(units=64,activation='relu')(shared_layer2)

# Create the output layer
att_output=layers.Dense(2,activation='sigmoid',name='att_output')(att_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dep_output, att_output])

# Compile the model
model.compile(
    optimizer='adam',
    loss=['categorical_crossentropy', 'binary_crossentropy'],
    metrics=['accuracy','accuracy']
)

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    [y_train_dep_encoded, y_train_att_encoded],
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
loss, department_accuracy, attrition_accuracy = model.evaluate(
    X_test_scaled,
    [y_test_dep_encoded, y_test_att_encoded]
)
```

```python
# Print the accuracy for both department and attrition
print(f'Department Accuracy: {department_accuracy:.4f}')
print(f'Attrition Accuracy: {attrition_accuracy:.4f}')
```
